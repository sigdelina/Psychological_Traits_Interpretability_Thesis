```python
def walk_tree(node, depth):
    if node.n_lefts + node.n_rights > 0:
        return max(walk_tree(child, depth + 1) for child in node.children)
    else:
        return depth


print([walk_tree(sent.root, 0) for sent in doc.sents])
```

```python
import numpy as np
from tqdm.auto import tqdm
```

```python
def dependency_extraction(text):
  
    list_max_depthes = []
    relative_clauses = {'acl:relcl':0, 'acl':0, 'advcl':0}


    doc = nlp(text)
    for sent in doc.sents:
      list_max_depthes.append(walk_tree(sent.root, 0))

      for token in sent:
        if token.dep_ in relative_clauses.keys():
          relative_clauses[token.dep_] += 1
    
    return np.round(np.mean(list_max_depthes), 1), max(list_max_depthes), min(list_max_depthes), relative_clauses
```

```python
from collections import defaultdict
```

```python
mean_depth, max_depth, min_depth = [], [], []
default_relative_clauses = defaultdict(list)

for ind, row in tqdm(df.iterrows()):
  mean_d, max_d, min_d, relativ_claues = dependency_extraction(row.punctuated)
  mean_depth.append(mean_d)
  max_depth.append(max_d)
  min_depth.append(min_d)
  for k, v in relativ_claues.items():
    default_relative_clauses[k].append(v)
```

```python
df = df.join(pd.DataFrame(
    {
        'mean_sent_depth': mean_depth,
        'max_synt_depth': max_depth,
        'min_synt_depth': min_depth,
        'N_acl:relcl':default_relative_clauses['acl:relcl'], 
        'N_acl':default_relative_clauses['acl'],
        'N_advcl':default_relative_clauses['advcl'],
    }, index=df.index
))
```

```python
df['N_relative'] = df.apply(lambda x: x['N_acl:relcl'] + x['N_acl'] + x['N_advcl'], axis=1)
```

```python
def morphological_feat_extraction(text):

    morphol = {
               'N_pronn_pers_first':0,
               'N_pronn_pers_sec_third':0,
               'N_adj':0,
               'N_adj_degree_compar':0, 
               'N_adj_degree_super':0,
               'N_verb_mood_ind':0,
               'N_verb_mood_imp':0,
               'N_verb_aspect_imp':0,
               'N_verb_aspect_perf':0,
               'N_verb_voice_act':0, 
               'N_verb_voice_pass':0,
               'N_verb_voice_mid':0,
               'N_verb_tense_past':0,
               'N_verb_tense_pres':0, 
               'N_verb_tense_fut':0 
               }


    doc = nlp(text)
    for sent in doc.sents:
      for token in sent:

        if token.dep_ == 'ROOT':
          if token.tag_ == 'VERB':

            aspect = token.morph.get("Aspect")
            if aspect != [] and aspect[0] == 'Imp':
              morphol['N_verb_aspect_imp'] += 1
            elif aspect != [] and aspect[0] == 'Perf':
              morphol['N_verb_aspect_perf'] += 1
            

            mood = token.morph.get("Mood")
            if mood != [] and mood[0] == 'Ind':
              morphol['N_verb_mood_ind'] += 1
            elif mood != [] and mood[0] == 'Imp':
              morphol['N_verb_mood_imp'] += 1
            

            tense = token.morph.get("Tense")
            if tense != [] and tense[0] == 'Past':
              morphol['N_verb_tense_past'] += 1
            elif tense != [] and tense[0] == 'Pres':
              morphol['N_verb_tense_pres'] += 1
            elif tense != [] and tense[0] == 'Fut':
              morphol['N_verb_tense_fut'] += 1

            
            voice = token.morph.get("Voice")
            if voice != [] and voice[0] == 'Act':
              morphol['N_verb_voice_act'] += 1
            elif voice != [] and voice[0] == 'Pass':
              morphol['N_verb_voice_pass'] += 1
            elif voice != [] and voice[0] == 'Mid':
              morphol['N_verb_voice_mid'] += 1
        
        if token.tag_ == 'ADJ':
          morphol['N_adj'] += 1

          degree = token.morph.get("Degree")


          if degree != []:
            if degree[0] == 'Abs' or degree[0] == 'Sup':
              morphol['N_adj_degree_super'] += 1

            elif degree[0] == 'Cmp':
              morphol['N_adj_degree_compar'] += 1
        
        if token.pos_ == 'PRON':
          person = token.morph.get("Person")
          if person:
            if person[0] == 'First':
              morphol['N_pronn_pers_first'] += 1
            else:
                morphol['N_pronn_pers_sec_third'] += 1
    return morphol
```

```python
from collections import defaultdict
```

```python

morph_params = defaultdict(list)

for ind, row in tqdm(df.iterrows()):
    morphological_features = morphological_feat_extraction(row.punctuated)
    for k, v in morphological_features.items():
      morph_params[k].append(v)
```

Для частотности возьмем:



*   Средняя частотность текста по IPM
*   Средняя частотность текста по D
*   Частотность самого редкого слова по IPM, D




```python

mystem = Mystem()

def freq_extraction(text):
  
    freqs_ipm, freqs_d  = [], []

    doc = nlp(text)

    for token in doc:
      analysis = mystem.analyze(token.text)[0]
      if 'analysis' in analysis:
        if analysis['analysis'] != []:
          lemma = analysis['analysis'][0]['lex']
          pos = analysis['analysis'][0]['gr'].split(',')[0].split('=')[0].lower()
          finding_freq = freq_vocab[(freq_vocab['Lemma'] == lemma) & (freq_vocab['PoS'] == pos)]
          try:
            freqs_ipm.append(finding_freq['Freq(ipm)'].tolist()[0])
            freqs_d.append(finding_freq['D'].tolist()[0])
          except:
            pass
    
    if freqs_ipm != [] and freqs_d != []:
        return {'mean_ipm': np.mean(freqs_ipm),
                'mean_d': np.mean(freqs_d),
                'min_freq_ipm': min(freqs_ipm),
                'min_freq_d': min(freqs_d)}
    else:
        return {'mean_ipm': np.NaN,
                'mean_d': np.NaN,
                'min_freq_ipm': np.NaN,
                'min_freq_d': np.NaN}
```

## Statistical features 

1) Number of words in the essay

2) Average length of a sentence in the essay

3) Length of the longest sentence in the essay

4) Average length of word in the essay

5) Length of the longest word in the essay

6) Number of repetitions of words used in the essay. The word most frequently repeated.

```python
from razdel import tokenize
from collections import Counter
```

```python
def stat_extraction(row):

    def remove_tokens(token):
      punct = ',.?!)(_;:'
      for punc in punct:
        token = token.replace(punc, '')
      return token

    cleaned_text = row.text
    punctuated_text = row.punctuated

    doc = nlp(punctuated_text)
    raw_doc = nlp(cleaned_text)

    
    # list of sentences splitted for tokens
    sentences = [sent.text.split(' ') for sent in doc.sents]

    # tokens from list of sentences for overall doc
    tokens = [token for subsents in sentences for token in subsents]

    # number of tokens in for each sentence
    tokens_param_sent = [len(toks) for toks in sentences]

    # length and lemmas for each token in a doc
    tokens_param_word = [(len(remove_tokens(token.text)), token.lemma_) for token in raw_doc]

    # length of each token
    words_length = [single_tok[0] for single_tok in tokens_param_word]

    # counting the existence of each token in a doc
    token_counter = Counter([single_tok[1] for single_tok in tokens_param_word])
    
    # for tokens with more than 1 existence value
    repeated_dict = dict((k, v) for k, v in  dict(token_counter).items() if v > 1)
    
    return {'N_words': len(tokens),
            'mean_Nwords_sent': np.mean(tokens_param_sent),
            'max_Nwords_sent': max(tokens_param_sent),
            'mean_len_words': np.mean(words_length),
            'max_len_words': max(words_length), 
            'N_repeated_words': 0 if not repeated_dict else sum(repeated_dict.values()),
            'N_most_repeated_word': 0 if not repeated_dict else max(repeated_dict.values())
            }
```

```python
stat_params = defaultdict(list)

for ind, row in tqdm(df.head(10).iterrows()):
    statistical_features = stat_extraction(row) 
    for k, v in statistical_features.items():
      stat_params[k].append(v)
```

```python
df = df.join(pd.DataFrame(
            {
            'N_words': stat_params['N_words'],
            'mean_Nwords_sent': stat_params['mean_Nwords_sent'],
            'max_Nwords_sent': stat_params['max_Nwords_sent'],
            'mean_len_words': stat_params['mean_len_words'],
            'max_len_words': stat_params['max_len_words'], 
            'N_repeated_words': stat_params['N_repeated_words'],
            'N_most_repeated_word': stat_params['N_most_repeated_word']
    }, index=df.index
))
```
